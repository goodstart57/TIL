```python
from bigsdatachatbot.utils import stock_datetime as sdt
from bigsdatachatbot import all_stocks_updater as as_updater
from datetime import datetime

import FinanceDataReader as fdr
import pandas as pd

_holiday = sdt.get_stock_holiday('C:/Users/student/Downloads')
# 모든 종목 코드와 종목명 수집
cn = as_updater.collect_all_stocks_cn()
# 오늘 날짜, 개장일 기준으로 60일 리스트 수집
start = sdt.get_n_open_date(today=datetime.now(), n=60, holiday=_holiday)
# 가격을 수집할 시작날짜, 끝 날짜 저장
start, end = min(start).strftime("%Y-%m-%d"), max(start).strftime("%Y-%m-%d")
all_stocks_price, passed_codes = [], []
```

```python
for code in tqdm(cn['code'].tolist()):
    try:
        price_temp = fdr.DataReader(code[1:] if code[0]=='a' else code, start, end)
        price_temp = price_temp[['Date', 'Close']].copy()
        price_temp.loc[:, 'code'] = code
        price_temp.index = range(price_temp.shape[0])
        all_stocks_price.append(price_temp)
    except:
        passed_codes.append(code)
all_stocks_price = pd.concat(all_stocks_price)
all_stocks_price.columns = list(map(lambda x: x.lower(), all_stocks_price.columns))
# save AllStocksPrice
```
